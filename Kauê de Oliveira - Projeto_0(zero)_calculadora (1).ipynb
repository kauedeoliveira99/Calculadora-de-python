```python
# Inicio da calculadora
print('---CALCULADORA DE PITONS RISOS(RS)---')

num1 = float(input('Digite um valor:'))
num2 = float(input('Digite o próximo valor:'))

# Adicionando as operações
print('\nEscolha a operação:')
print('1) adição +')
print('2) subtração -')
print('3) multiplicação *')
print('4) divisão /')

operação = (input('Digite o número da operação desejada:'))

# Aplicando as operações
print('\n---RESULTADO---')

if operação == '+':
  resultado = num1 + num2
  print(f'{num1} + {num2} = {resultado}')

elif operação == '-':
    resultado = num1 - num2
    print(f'{num1} - {num2} = {resultado}')

elif operação == '*':
    resultado = num1 * num2
    print(f'{num1} * {num2} = {resultado}')

elif operação == '/':
    if num2 == 0:
      print('Erro: não é possivel dividir por zero!')
    else:
      resultado = num1 / num2
      print(f'{num1} / {num2} = {resultado}')

else:
  print('Erro: Operação inválida!')
```
